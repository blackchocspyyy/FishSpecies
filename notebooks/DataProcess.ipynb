```python
unique_fish = analysis_data['fishNum'].unique()

# Convert to a list
csv_names = [f"{fish}.csv" for fish in unique_fish]

# Display the list of CSV file names
print(csv_names)
```

```python
# Define the fishes to exclude
# Those are inactive fishes based on the fishNotes_20220912.csv
exclude_list = ['LT013', 'LT015', 'LT019', 'SMB004', 'SMB010','SMB013']

# Filter out entries starting with 'SMB' and the specified fishes
filtered_fish = analysis_data[~analysis_data['fishNum'].str.startswith('LWF')]
filtered_fish = filtered_fish[~filtered_fish['fishNum'].isin(exclude_list)]

# Get the list of unique fish names for the filtered dataset
unique_fish = filtered_fish['fishNum'].unique()

# Convert to a list of CSV names
csv_names = [f"{fish}.csv" for fish in unique_fish]

# Display the list of CSV file names
print(csv_names)
```

```python
csv_names
```

```python
# Regular expression to match columns like F45 or F45.5
pattern = r'^F\d+(\.\d+)?$'

# List to store datasets with no missing values
datasets_no_missing = []

# Loop through each dataset
for dataset in csv_names:
    try:
        # Load the dataset
        df = pd.read_csv(dataset)

        # Filter columns that match the pattern
        f_columns = [col for col in df.columns if re.match(pattern, col)]

        # Check if there are any missing values in these columns
        if f_columns:  # Ensure there are columns matching the pattern
            missing_values = df[f_columns].isna().sum()
            
            # If no missing values in any of the F columns
            if missing_values.sum() == 0:
                datasets_no_missing.append(dataset)

            # Display missing values for this dataset
            print(f"Dataset: {dataset}")
            print(f"Missing values in columns starting with 'F' followed by numbers:\n{missing_values}")
            print("-" * 40)
        else:
            print(f"Dataset: {dataset} - No columns starting with 'F' followed by numbers found.")
    except FileNotFoundError:
        print(f"File {dataset} not found.")

# Print the list of datasets with no missing values
print("\nDatasets with no missing values in frequency columns:")
print(datasets_no_missing)
```

```python
rows = 0
for dataset in datasets_no_missing:
    df = pd.read_csv(dataset)
    rows+=len(df)
    print(f"{dataset}: {len(df)} rows")
```
